```python
#Project QuickDraw Doodle Recognition
#File - Model Building and Testing

#  About:
#  This code uses the created model by loading the weights
#  that we have generated using MobileNet, saved under model directory.

#Import essential libraries

import pandas as pd
import numpy as np
import seaborn as sns
import cv2
import os
#import warnings to suppress warnings
import warnings
warnings.filterwarnings("ignore")
import ast
import math
import matplotlib.pyplot as plot
from simplification.cutil import simplify_coords

#import keras libraries
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.applications.mobilenet import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow import keras

from keras.models import load_model
from tensorflow.keras.metrics import categorical_accuracy, top_k_categorical_accuracy, categorical_crossentropy


N = 10 #Number of Categories for doodle
total_number_of_csvs = 10
train_csvs = 8
test_csvs = 1
small_image_size = 64
original_image_size = 256
batchsize = 680
train_rec_count = 0

input_dir_path = 'data/'
shuffled_path = 'data/shuffled_data/'
TRAIN_DIR = 'data/train_simplified/'
model_name = './mobile_net_v2.model'


def list_doodles_to_train():
    files = os.listdir(TRAIN_DIR)
    if '.DS_Store' in files:
        files.remove('.DS_Store')
    return sorted([(fil.split('.')[0]) for fil in files], key = str.lower)

def top_two_accuracy(y_true, y_pred):
    return top_k_categorical_accuracy(y_true, y_pred, k=2)

def pretty_print(lst):
    count = 0
    for doodle in lst:
        print('Doodle Category is represented as ', (doodle,count))
        count+=1
        
def plot_me(x,n):
    fig, axs = plot.subplots(nrows=n, ncols=n, sharex=True, sharey=True, figsize=(12, 12))
    for i in range(n**2):
        ax = axs[i // n, i % n]
        (-x[i]+1)/2
        ax.imshow((-x[i, :, :, 0] + 1)/2, cmap=plot.cm.gray)
        ax.axis('off')
    plot.tight_layout()
    plot.show();
    
def process_x(chunk,small_image_size):
    x = np.zeros((len(chunk),small_image_size,small_image_size,1))
    chunk['drawing'] = chunk['drawing'].apply(ast.literal_eval)
    for i,stroke_array in enumerate(chunk.drawing.values):
        x[i,:,:,0] = reshape_doodle_get_array(stroke_array, small_image_size)
    return preprocess_input(x).astype(np.float32)

def get_validation_data():
    validation_data = pd.read_csv(os.path.join(shuffled_path, 'chunk{}.csv.gz'.format(8)))
    xv = process_x(validation_data, small_image_size)
    yv = keras.utils.to_categorical(validation_data.y, num_classes=N)
    print('-------------------Validation data ----------------------------')
    print('Shape of X',xv.shape)
    print('Shape of y',yv.shape)
    print('---------------------------------------------------------------')
    return xv,yv,xv.shape[0]

def get_test_data():
    test_data = pd.read_csv(os.path.join(shuffled_path, 'chunk{}.csv.gz'.format(9)))
    xt = process_x(test_data,small_image_size)
    yt = test_data['word']
    print('-------------------Test data Shape------------------------------')
    print('Shape of X',xt.shape)
    print('Shape of y',yt.shape)
    print('---------------------------------------------------------------')
    return xt,yt,xt.shape[0]

def reshape_doodle_get_array(strokes_for_doodle, small_image_size, width_line = 5):
    image = np.zeros((original_image_size, original_image_size), np.uint8)
    for i, one_stroke_with_x_y in enumerate(strokes_for_doodle):
        for i in range(len(one_stroke_with_x_y[1]) - 1): #get number of points in one stroke
            #one_stroke_with_x_y = [[All X],[All Y]] so [0] is x and [1] is y
            image = cv2.line(image, (one_stroke_with_x_y[0][i], one_stroke_with_x_y[1][i]),
                         (one_stroke_with_x_y[0][i + 1], one_stroke_with_x_y[1][i + 1]), 230, width_line) # 0 re
    return cv2.resize(image, (small_image_size, small_image_size))

def process_image_for_mobilenet(scaled_down_size,batchsize,train_csv_count,width_line = 5):
    while True:
        for i in np.random.permutation(train_csv_count):
            input_csv = os.path.join(shuffled_path,'chunk{}.csv.gz'.format(i))
            for chunk in pd.read_csv(input_csv, chunksize = batchsize):
                   #For every chunk send the x and y values to the original functio
                global train_rec_count
                train_rec_count = train_rec_count + chunk.shape[0]
                x = process_x(chunk,small_image_size)
                y = keras.utils.to_categorical(chunk.y, num_classes=N)
                yield x, y

def preds2catids(predictions):
    return pd.DataFrame(np.argsort(-predictions, axis=1)[:, :2], columns=['guess_1', 'guess_2'])

cats = list_doodles_to_train()
print('---------------------------------------------------------------------------')
pretty_print(cats)
```

```python
model = MobileNet(input_shape=(64, 64, 1), alpha=1., weights=None, classes=10)
model.load_weights('model/my_model_v2.h5')
print('--------------------------Model Summary-------------------------------------')
print(model.summary())
model.compile(optimizer=Adam(lr=0.002), loss='categorical_crossentropy',
              metrics=[categorical_crossentropy, categorical_accuracy, top_two_accuracy])
```

```python
print('----------------------------DATA EXPLORATION-------------------------------')
train_explore = pd.DataFrame()
for category in cats:
    train_explore = train_explore.append(pd.read_csv(TRAIN_DIR + category +'.csv'))
print('\nNature of the Input Data:\n')
print(train_explore.head())

count_records=[0] * 10
for category in train_explore['word']:
    count_records[cats.index(category)] = count_records[cats.index(category)]+1
print('---------------------------------------------------------------------------')
print("Total Number of records before splitting the data into train,test,validation")
print(train_explore.shape)
total_train_count = train_explore.shape[0]
print('---------------------------------------------------------------------------')
print('Plotting the number records present in each category: ')
pd.options.display.max_rows = 20
sns.set(style="darkgrid")
plot.figure()
ax_t10 = sns.barplot(x=cats, y=count_records, palette="coolwarm")
ax_t10.set_xticklabels(ax_t10.get_xticklabels())
plot.show() 
```

```python
xv,yv,val_count = get_validation_data()
plot_me(xv,7)
```

```python
xt,yt,test_count = get_test_data()
plot_me(xt,10) 
```

```python
print('-------------------------------Training--------------------------------------------')

print("Total Number of Training Samples")
print(total_train_count - val_count - test_count )
training_data = process_image_for_mobilenet(scaled_down_size=small_image_size, batchsize=batchsize, train_csv_count=range(train_csvs - 1))
x, y = next(training_data)
plot_me(x,13)
```

```python
test_predictions = model.predict(xt, batch_size=128, verbose=1)
```

```python
top2 = preds2catids(test_predictions)
top2.head()
top2.shape
    
cats = list_doodles_to_train()
id2cat = {k: cat.replace(' ', '_') for k, cat in enumerate(cats)}
top2guesses = top2.replace(id2cat)
top2guesses.head()
top2guesses.shape
```

```python
top2guesses['Actual_doodle'] = yt
print(top2guesses)
```

```python
#find testing accuracy based on guess_1
from sklearn.metrics import accuracy_score
print('Testing Accuracy = \n Guess1: ')
guess1_accuracy = accuracy_score(top2guesses['Actual_doodle'],top2guesses['guess_1'])
print(guess1_accuracy)
print('\n Guess2: ')
guess2_accuracy = accuracy_score(top2guesses['Actual_doodle'],top2guesses['guess_2'])
print(guess2_accuracy)
print('Total Test Accuracy = ',(guess1_accuracy+guess2_accuracy) )
```
